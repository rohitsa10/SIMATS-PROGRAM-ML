LAB EXPERIMENT 1: Implementation of Bayes Theorem for Classification

```python
# Bayes Theorem based classification
p_yes = 0.6
p_no = 0.4
p_x_given_yes = 0.7
p_x_given_no = 0.2
p_yes_given_x = (p_x_given_yes * p_yes) / (
p_x_given_yes * p_yes + p_x_given_no * p_no)
p_no_given_x = 1 - p_yes_given_x
print("Posterior Probability (Yes):", p_yes_given_x)
print("Posterior Probability (No):", p_no_given_x)
```

LAB EXPERIMENT 2: Maximum Likelihood Estimation (MLE)

```python
import numpy as np
# Sample data
X = np.array([4, 5, 6, 5, 4, 6, 5])
mean_mle = np.mean(X)
var_mle = np.var(X)
print("Estimated Mean:", mean_mle)

print("Estimated Variance:", var_mle)
```

LAB EXPERIMENT 3: Naïve Bayes Classifier (From Scratch)

```python
import numpy as np
X = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array(['Yes','Yes','No','No'])
classes = np.unique(y)
priors = {}
likelihood = {}
for c in classes:
    X_c = X[y == c]
    priors[c] = len(X_c) / len(X)
    likelihood[c] = np.mean(X_c, axis=0)
x_test = np.array([1,1])
posteriors = {}
for c in classes:
    posteriors[c] = priors[c] * np.prod(likelihood[c] ** x_test)
print("Prediction:", max(posteriors, key=posteriors.get))
```

LAB EXPERIMENT 4: Naïve Bayes Classification using Scikit-learn

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

LAB EXPERIMENT 5: Expectation Maximization (EM) Algorithm using GMM

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=200, centers=3, random_state=42)
model = GaussianMixture(n_components=3)
model.fit(X)
labels = model.predict(X)

print("Cluster Labels:", labels[:10])
```
